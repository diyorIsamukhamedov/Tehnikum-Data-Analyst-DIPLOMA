## **2**. Переименовать колонки и убрать символ переноса строки \n из названий, например: "Номер \nзаказа" -> "Номер заказов".
---
### Были обнаружены орфографические и смысловые ошибки в названиях столбцов:

| Оригинал | Рекомендованное исправление | Ожидаемый результат |
| ----------- | ----------- | ----------- |
| Номер заказа | Номер заказов | Номер заказов |
| Указаное дата доставки | Указанная дата доставки | ❌ Ошибка: указаное → указанная |
| Указаное время доставки | Указанное время доставки | ❌ Ошибка: указаное → указанное |
| Стоимость товаров после формирование, сумы | Стоимость товаров после формирования, сум | ❌ формирование → формирования, и сумы → сум |
| Сумма по чеку, сумы | Сумма по чеку, сум | ❌ сумы → сум |
| Минимальная сумма доставки, сумы | Минимальная сумма доставки, сум | ❌ сумы → сум |
| Оплата через систему Octo (sum) | Оплата через систему Octo (сум) | ❌ sum → сум (русский язык) |
| Всего оплачено с картой | Всего оплачено с карты | ✔ Логичнее: оплачено с карты, а не с картой |
| Id Сборщика | ID Сборщика | ⚠️ Id → ID (принято писать в верхнем регистре) |
---

```python
# Импортируем модуль регулярных выражений для работы с шаблонами замены
import re

# Шаг 1: Очистка от \n и лишних пробелов
# .rename() — метод, который позволяет переименовать столбцы и индексы в DataFrame
# columns = ... — указывает, что мы хотим переименовать столбцы
# lambda x: ... — это анонимная функция, применяемая к каждому названию столбца
# re.sub(r"\s+", " ", x) — заменяет все подряд идущие пробельные символы (включая пробелы, \n, табуляции) на один пробел
# .sub() — функция из модуля re, которая заменяет все вхождения шаблона на заданную строку
# r"\s+" - буква r перед строкой указывает на то, что это "сырая" строка, в которой обратные слеши не экранируются
# \s — соответствует любому пробельному символу (пробел, табуляция, перенос строки и т.д.)
# + — означает "один или более" (то есть, если пробельных символов несколько подряд, они будут заменены на один пробел)
# .strip() — удаляет пробелы в начале и в конце строки
# inplace=True — означает, что изменения будут внесены непосредственно в исходный DataFrame df_excel
# Результат — чистые и единообразные названия столбцов без лишних пробелов и переносов
df_excel.rename(columns = lambda x: re.sub(r"\s+", " ", x).strip(), inplace = True)

# Шаг 2: Переименование столбцов на корректные имена
corrected_columns = {
    "Номер заказа": "Номер заказов",
    "Указаное дата доставки": "Указанная дата доставки",
    "Указаное время доставки": "Указанное время доставки",
    "Стоимость товаров после формирование, сумы": "Стоимость товаров после формирования, сум",
    "Сумма по чеку, сумы": "Сумма по чеку, сум",
    "Минимальная сумма доставки, сумы": "Минимальная сумма доставки, сум",
    "Оплата через систему Octo (sum)": "Оплата через систему Octo (сум)",
    "Всего оплачено с картой": "Всего оплачено с карты",
    "Id Сборщика": "ID Сборщика"
}

df_excel.rename(columns = corrected_columns, inplace = True)

# Показываем имена столбцов
df_excel.columns
```

## **3**. Все ID привести к формату int (например, заменить 1.0 на 1).
---

```python
# Проходим по всем названиям столбцов в таблице df_excel
for col in df_excel.columns:
    # Проверяем, содержит ли название столбца подстроку "ID"
    # Это позволяет выбрать все столбцы, которые являются идентификаторами
    if "ID" in col:
        # Преобразуем тип данных в Int64 (nullable integer):
        # - позволяет работать с целыми числами
        # - сохраняет пропущенные значения (NaN)
        # - устраняет дробную часть у чисел вроде 1.0 → 1
        df_excel[col] = df_excel[col].astype("Int64")

# Проверяем, что все столбцы с ID имеют тип Int64
df_excel[["ID Доставщика", "ID Сборщика"]].dtypes
```

```python
# Получаем название последнего столбца в датафрейме ("Номер адреса доставки")
last_col = df_excel.columns[-1]

# Преобразование последней колонки в целые числа БЕЗ научной нотации:
# .fillna(0) — заменяет NaN на 0
# .round(0) — если есть float с дробной частью, делает округление
# .astype(int) — приводит к int (бросит ошибку, если остаются нецелые значения)
df_excel[last_col] = df_excel[last_col].fillna(0).round(0).astype(int)

# Проверяем результат преобразования: первые 10 строк датафрейма последней строки "Номер адреса доставки"
df_excel[[last_col]].head(10)
```

## **5**. Перевести все колонки с типом данных времени на формат `"datetime"`.
---
### Результат анализа всех потенциальных колонок с типом данных "время" по фактическому содержимому:
| Название колонки | Ожидаемый результат |
| ----------- | ----------- |
| Дата создания заказа | ✅ Содержит дату. Преобразовать к `datetime`. |
| Указанная дата доставки | ✅ Содержит дату. Преобразовать к `datetime`. |
| Время создания заказа | ✅ Содержит только время. Преобразовать в `datetime`. |
| Время приема заказа | ✅ Содержит только время. Преобразовать в `datetime`.|
| Указанное время доставки | ✅ Содержит только время. Преобразовать в `datetime`. |
| Фактическое время доставки | ✅ Содержит только время. Преобразовать в `datetime`. |
| Время загрузки чека | ✅ Преобразовать к `datetime` с форматом %d-%m-%Y %H:%M. |

### ⚠️ При приведении столбцов с временем (например: **"12:45"**) к формату datetime, pandas автоматически добавляет дату по умолчанию (**1900-01-01**), так как тип данных datetime64[ns] требует наличия и даты, и времени.

### Удалить эту дату, оставив при этом тип `datetime64`, невозможно — в этом случае придётся использовать тип datetime.time, который в pandas хранится как `object`, что противоречит требованию "преобразовать в `datetime`".

### Поэтому оставляем тип `datetime64[ns]` для дальнейшего анализа, а для корректного визуального отображения времени — используем метод `.dt.time`.

### Финальная очистка отображения возможна при экспорте в **Excel** (например, через форматирование колонок).
---

```python
# 1. Указываем группы столбцов по содержанию
# Колонки, содержащие только ДАТУ в формате "dd-mm-YYYY"
date_only_columns = [
    "Дата создания заказа",
    "Указанная дата доставки"
]

# Колонки, содержащие только ВРЕМЯ в формате "HH:MM"
# Преобразовать в datetime, где дата будет по дэфолту в формате(1900-01-01)
time_only_columns = [
    "Время создания заказа",
    "Время приема заказа",
    "Указанное время доставки",
    "Фактическое время доставки"
]

# Колонка с датой и временем — формат: "dd-mm-YYYY HH:MM"
datetime_columns = [
    "Время загрузки чека"
]

# 2. Преобразуем столбцы с датами
# Используем формат "%d-%m-%Y" — день-месяц-год
# Указываем dayfirst=True, чтобы Pandas правильно понял день и месяц
for col in date_only_columns:
    df_excel[col] = pd.to_datetime(
        df_excel[col],
        format = "%d-%m-%Y",  # Пример: "31-12-2021"
        errors = "coerce",    # если значение не удаётся преобразовать, то → NaT
        dayfirst = True       # указываем, что сначала идёт день (а не месяц)
    )

# 3. Преобразуем столбцы с временем
# Используем формат "%H:%M" — часы и минуты (например, 12:45)
# Дата при этом будет по умолчанию в формате — 1900-01-01
for col in time_only_columns:
    df_excel[col] = pd.to_datetime(
        df_excel[col],
        format = "%H:%M",    # Пример: "12-45" → 1900-01-01 12:45:00
        errors = "coerce"    # если значение не удаётся преобразовать, то → NaT
    ).dt.time  # оставляем только время (без даты)

# 4. Преобразуем столбцы с датой и временем
# Используем формат "%d-%m-%Y %H:%M"
# Например: "31-12-2021 14:45"
for col in datetime_columns:
    df_excel[col] = pd.to_datetime(
        df_excel[col],
        format = "%d-%m-%Y %H:%M",  # Пример: "31-12-2021 14:30"
        errors = "coerce",          # если значение не удаётся преобразовать, то → NaT
        dayfirst = True             # указываем, что сначала идёт день (а не месяц)
    )

# 5. Проверяем результат преобразования: первые 5 строк датафрейма
df_excel[date_only_columns + time_only_columns + datetime_columns].head()
```

### **7**. `DataFrame.describe()` - используется для получения сводной статистики по DataFrame. Он автоматически рассчитывает количество ненулевых значений (count), среднее значение (mean), стандартное отклонение (std / разброс данных), минимальное значение (min), 25-й, 50-й (медиана) и 75-й процентили (quartiles), а также максимальное значение (max) для каждого числового столбца в DataFrame.
---
### **8**. `DataFrame.info()` - функция `info()` показывает информацию о DataFrame. Включая типы данных и использование памяти memory_usage = "deep" позволяет получить более точную информацию о потреблении памяти. Важно отметить, что использование memory_usage = "deep" может занять больше времени, так как оно требует дополнительного анализа.
---
### **9**. `DataFrame.rename() / Series.rename()` - используется для переименования столбцов или индексов в DataFrame и Series.
---
### **10**. `DataFrame.astype()` - изменяет тип данных столбца.
---
### **11**. `DataFrame.fillna()` - зполняет пропущенные значения (NaN) определённым значением.
---
### **12**. `pd.to_datetime()` - преобразует строку или столбец в формат даты/времени datetime64.
---
### **13**. `DataFrame.groupby()` - группирует данные по значению столбца (как GROUP BY в SQL). Схлопывает данные.
---
### **14**. `DataFrame.count()` - возвращает кол-во непустых (не NaN) значений по столбцам или строкам.
---
### **15**. `DataFrame.reset_index()` - сброс индекса в DataFrame, полезно после groupby().
---
### **16**. `DataFrame.loc[]` - метод `loc` в Pandas используется для **выборки данных** на основе меток (названий) строк и столбцов.
---
### **17**. `DataFrame.iloc[]` - метод `iloc` используется для выборки данных на основе позиций (номеров) строк и столбцов.
---
### **18**. `DataFrame['col'].unique()` - возвращает уникальные значения в столбце.
---
### **19**. `len(df)` - возвращает количество строк (как COUNT(*) в SQL).
---
### **20**. `DataFrame.agg()` - позволяет выполнять несколько агрегатных функций сразу.
---
### **21**. `DataFrame.merge(..., how='inner/left/right/outer')` - Объединяет два DataFrame по ключу, как JOIN в SQL. inner — только совпадающие строки, left — все строки из левого + совпадающие, right — из правого, outer — всё (аналог FULL JOIN).
---
### **22**. `DataFrame.pivot_table()` - Создаёт сводную таблицу похожую в Excel, поддерживает агрегатные функции (sum, mean и др.).
---
### **23**. `DataFrame.pivot()` - создаёт таблицу, где значения становятся строками и столбцами (без агрегации).
---
### **24**. `np.random.randint()` - генерирует случайные целые числа в заданном диапазоне.
---
### **25**. `DataFrame.sample() / Series.sample()` - используется для случайной выборки строк или элементов из DataFrame или Series.
